### Handling Missing Value

it show the steps to hanlding missing value

```python
### courier unique check 

amazon['Courier Status'].unique()
```

This data value is not a numerical type and its about the status shipment, so i will drop missing value from the courier status columns

```python
amazon.dropna(subset=['Courier Status'], inplace = True)
```

```python
### currency unique check

amazon['currency'].unique()
```

currency unique : 
- INR
- nan

by the domain knowledge, *nan* will replace with *INR*. INR is indian currency. from data extraction shows that it is the transaction sales to India.

```python
### change the nan value on currency

amazon['currency'] = amazon['currency'].fillna('INR')
```

```python
### Check unique Amount

amazon['Amount'].unique()
```

this is numerical data, need deep analyzing like :
- Price the similar goods
- city origin of delivery goods
- Unit Price
- Delivery price for each unit

by domain knowledge, we cant found the detail information of each unit/product and the information obtained is biased, so it can influence the results of the analysis

we cant fill the missing value by the median, it is feared that predictions will be wrong, such as commodity prices missing so that the analysis results will not be as good as possible

it will be drop missing value

```python
### drop missing value on Amount columns
amazon.dropna(subset=['Amount'], inplace = True)
```

```python
### Check unique ship-city

amazon['ship-city'].unique()
```

we will drop the missing value

```python
### drop missing value ship-city

amazon.dropna(subset=['ship-city'], inplace = True)
```

### Cek Outliers

```python
amazon.describe() ## describe the numerical data
```

```python
q1 = amazon.describe()['Amount']['25%']
q3 = amazon.describe()['Amount']['75%']
iqr = q3-q1
upper_f = q3 + (1.5 * iqr)
lower_f = q1 - (1.5 * iqr)
outliers = amazon[(amazon['Amount'] < lower_f) | (amazon['Amount'] > upper_f)]
outliers
```

```python
import matplotlib.pyplot as plt
```

```python
plt.boxplot(amazon['Amount'])
plt.show()
```
